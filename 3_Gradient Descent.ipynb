## λ―Έλ¶„

- λ―Έλ¶„(differentiation)μ€ **λ³€μμ μ›€μ§μ„μ— λ”°λ¥Έ ν•¨μκ°’μ λ³€ν™”λ¥Ό μΈ΅μ •ν•κΈ° μ„ν• λ„κµ¬**λ΅ μµμ ν™”μ—μ„ μ μΌ λ§μ΄ μ‚¬μ©ν•λ” κΈ°λ²•μ…λ‹λ‹¤.

- λ―Έλ¶„μ€ λ³€ν™”μ¨(κΈ°μΈκΈ°) μ κ·Ήν•(limit)μΌλ΅ μ •μ, h λ¥Ό 0 μΌλ΅ λ³΄λ‚΄λ©΄ (x, f(x))μ—μ„ μ ‘μ„ μ κΈ°μΈκΈ°λ΅ μλ ΄ν•λ‹¤.
    
    - $ f'(x) = \lim\limits_{h\to0}\dfrac{f(x+h)-f(x)}{h} $

- `sympy.diff` λ¥Ό κ°€μ§€κ³  λ―Έλ¶„ κ³„μ‚° κ°€λ¥


```python
# sympy : symbolic python, 
# poly : polynomial(λ‹¤ν•­μ‹)
# xμ— λ€ν•΄μ„ λ―Έλ¶„

import sympy as sym
from sympy.abc import x

 # Poly(2*x + 2, x, domain='ZZ')
sym.diff(sym.poly(x**2 + 2*x + 3), x) 
```

## λ―Έλ¶„ μ΄ν•΄

- λ―Έλ¶„μ€ ν•¨μ f μ **μ£Όμ–΄μ§„ μ  (x, f(x)) μ—μ„μ, μ ‘μ„ μ κΈ°μΈκΈ°**λ¥Ό κµ¬ν•λ‹¤.
    -  λ―Έλ¶„μ„ κ³„μ‚°ν•λ ¤λ©΄ ν•¨μμ λ¨μ–‘μ΄ λ§¤λ„λ¬μ›μ•Ό(μ—°μ†) ν•λ‹¤.


- ν• μ μ—μ„ μ ‘μ„ μ κΈ°μΈκΈ°λ¥Ό μ•λ©΄, μ–΄λ λ°©ν–¥μΌλ΅ μ μ„ μ›€μ§μ—¬μ•Ό ν•¨μκ°’μ΄ μ¦κ°€ν•λ”μ§€ / κ°μ†ν•λ”μ§€ μ• μ μμµλ‹λ‹¤.

    
    - ν•¨μλ¥Ό μ¦κ°€ μ‹ν‚¤κ³  μ‹¶μΌλ©΄, λ―Έλ¶„κ°’ f'(x) μ„ (μ–‘μ, μμ μƒκ΄€μ—†μ΄) λ”ν•λ©΄ λ©λ‹λ‹¤.
    
        - λ©μ ν•¨μλ¥Ό μµλ€ν™” ν•λ‹¤.
        - λ―Έλ¶„κ°’ f'(x) μ„ λ”ν•λ©΄, κ²½μ‚¬μƒμΉλ²•(gradient ascent)μ΄λΌ ν•λ©°, ν•¨μμ κ·Ήλ€κ°’μ μ„μΉλ¥Ό κµ¬ν•  λ• μ‚¬μ©ν•λ‹¤.

    - ν•¨μλ¥Ό κ°μ† μ‹ν‚¤κ³  μ‹¶μΌλ©΄, λ―Έλ¶„κ°’ f'(x) μ„ (μ–‘μ, μμ μƒκ΄€μ—†μ΄) λΉΌλ©΄ λ©λ‹λ‹¤.  
    
        - λ©μ ν•¨μλ¥Ό μµμ†ν™” ν•λ‹¤.
        - λ―Έλ¶„κ°’ f'(x) μ„ λΉΌλ©΄, κ²½μ‚¬ν•κ°•λ²•(gradient descent)μ΄λΌ ν•λ©°, ν•¨μμ κ·Ήμ†κ°’μ μ„μΉλ¥Ό κµ¬ν•  λ• μ‚¬μ©ν•λ‹¤.


- κ·Ήκ°’μ—μ„  λ―Έλ¶„κ°’μ΄ 0, f'(x) == 0 μ΄λ―€λ΅, κ²½μ‚¬μƒμΉ / κ²½μ‚¬ν•κ°• λ°©λ²•μ΄ λ” μ΄μƒ μ—°μ‚°(μ—…λ°μ΄νΈ)μ΄ μ•λλ―€λ΅, λ©μ ν•¨μ μµμ ν™”κ°€ μλ™μΌλ΅ μΆ…λ£ν•©λ‹λ‹¤.

## κ²½μ‚¬ ν•κ°•λ²• μ•κ³ λ¦¬μ¦

- gradient: λ―Έλ¶„μ„ κ³„μ‚°ν•λ” ν•¨μ
- init: μ‹μ‘μ 
- lr: ν•™μµλ¥ 
- eps: μ•κ³ λ¦¬μ¦ μΆ…λ£μ΅°κ±΄


```python
# λ³€μκ°€ ν•λ‚μΈ κ²½μ°, (x)

var = init

# gradient: λ―Έλ¶„μ„ κ³„μ‚°ν•λ” ν•¨μ
grad = gradient(var)

# 1-1) μ»΄ν“¨ν„°λ΅ κ³„μ‚°ν•  λ• λ―Έλ¶„μ΄ μ •ν™•ν 0μ΄ λλ” κ²ƒμ€ λ¶κ°€λ¥ν•λ―€λ΅, 
# eps(μ•κ³ λ¦¬μ¦ μΆ…λ£μ΅°κ±΄) λ³΄λ‹¤ μ‘μ„ λ•, μΆ…λ£ν•λ” μ΅°κ±΄μ΄ ν•„μ”ν•λ‹¤.
# λ―Έλ¶„κ°’μ μ λ€κ°’ μ΄μ©
while(abs(grad) > eps):
    
    # 1-2) κ²½μ‚¬ ν•κ°•λ²•, f'(x) λ¥Ό λΉΌλ” κ³Όμ •
    # `x β’ Ξ»fβ€²(x)` μ„ κ³„μ‚°ν•λ” λ¶€λ¶„. 
    # lr(ν•™μµλ¥ )μ€ ν•™μµλ¥ λ΅μ„, λ―Έλ¶„μ„ ν†µν•΄ μ—…λ°μ΄νΈν•λ” μ†λ„λ¥Ό μ΅°μ ν•λ‹¤. 
    var = var - (lr * grad)
    
    # 1-3) μΆ…λ£μ΅°κ±΄μ΄ μ„±λ¦½ν•κΈ° μ „κΉμ§€, λ―Έλ¶„κ°’μ„ κ³„μ† μ—…λ°μ΄νΈν•λ‹¤.
    grad = gradient(var)
```

## νΈλ―Έλ¶„(partial differentiation)

- λ―Έλ¶„(differentiation)μ€ **λ³€μμ μ›€μ§μ„μ— λ”°λ¥Έ, ν•¨μκ°’μ λ³€ν™”λ¥Ό μΈ΅μ • ν•κΈ° μ„ν• λ„κµ¬**λ΅ μµμ ν™”μ—μ„ μ μΌ λ§μ΄ μ‚¬μ©ν•λ” κΈ°λ²•μ…λ‹λ‹¤.

- μ…λ ¥(λ³€μ)κ°€ λ²΅ν„°μ™€ κ°™μ΄ λ‹¤λ³€μ ν•¨μμ κ²½μ°, νΈλ―Έλ¶„(partial differentiation)μ„ μ‚¬μ©ν•λ‹¤.

    - ν• λ°©ν–¥μ„ κΈ°μ¤€μΌλ΅λ§ μ²λ¦¬ x, xλ°©ν–¥ μ›€μ§μ„λ§ λ¶„μ„ μ²λ¦¬

    - $\partial_{xi}f(x) = \lim\limits_{h\to0}\dfrac{f(x+he_{i})-f(x)}{h} $
    
    - $e_{i}$ λ” i λ²μ§Έ κ°’λ§ 1μ΄κ³ , λ‚λ¨Έμ§€λ” 0μΈ λ‹¨μ„ λ²΅ν„°

---

- dμ°¨μ› κ° λ³€μ λ³„(x,y,z ..d)λ΅ νΈλ―Έλ¶„μ„ κ³„μ‚°ν• κ·Έλ λ””μ–ΈνΈ(gradient) λ²΅ν„°λ¥Ό μ΄μ©ν•μ—¬, λ‹¤λ³€μ μ…λ ¥ μ²λ¦¬, κ²½μ‚¬ν•κ°• / κ²½μ‚¬μƒμΉλ²•μ— μ‚¬μ©ν•  μ μλ‹¤.

- μ•μ„ μ‚¬μ©ν• λ―Έλ¶„κ°’μΈ $fβ€²(x)$ λ€μ‹  λ²΅ν„° $\nabla f$ (nabla)λ¥Ό μ‚¬μ©ν•μ—¬, λ³€μ x = (x1, ..., xd) λ¥Ό λ™μ‹μ— μ—…λ°μ΄νΈ κ°€λ¥ν•©λ‹λ‹¤.

    - $\nabla f = (\partial_{x1}f, \partial_{x2}f, ..., \partial_{xd}f)$


```python
# νΈλ―Έλ¶„ κ³„μ‚°

import sympy as sym

from sympy.abc import x, y

sym.diff(sym.poly(x**2 + 2*x*y + 3) + sym.cos(x + 2*y), x)

```

## κ·Έλ λ””μ–ΈνΈ λ²΅ν„°(gradient vector)


- κ° μ  (x, y, z) κ³µκ°„μ—μ„ ν•¨μ f(x, y) ν‘λ©΄μ„ λ”°λΌ β’β‡f (κ·Έλ λ””μ–ΈνΈ)λ²΅ν„° λ¥Ό λ“±κ³ μ„ (contour) ν•΄μ„
- μµμ ν™”
    
    - $f (x, y) = x^2 + 2y^2$

    - $\nabla f =(2x,4y)$
        - β‡f(x,y)λ” κ°μ (x,y)μ—μ„ κ°€μ¥ λΉ¨λ¦¬ μ¦κ°€ν•λ” λ°©ν–¥μΌλ΅ νλ¥΄κ² λλ‹¤.
        - κ·Ήλ€κ°’μ„ ν–¥ν•κ² ν•¨

    - $-\nabla f =-(2x,4y)$
        - β‡fλ” β‡(β’f)λ‘ κ°™κ³ , μ΄λ” κ° μ μ—μ„ κ°€μ¥ λΉ¨λ¦¬ κ°μ†ν•κ² λλ” λ°©ν–¥κ³Ό κ°™λ‹¤.
        - κ·Ήμ†κ°’μ„ ν–¥ν•κ² ν•¨

- κ²°λ΅ : λ³€μκ°€ λ²΅ν„°μΈ κ²½μ°, νΈλ―Έλ¶„μ„ ν†µν•΄μ„ κµ¬ν• κ·Έλλ””μ–ΈνΈ λ²΅ν„°λ¥Ό ν†µν•΄ d-μ°¨μ›μΌλ΅ κ²½μ‚¬ν•κ°•λ²•μ„ ν™•μ¥ν•  μ μλ‹¤.

```python
var = init

# - gradient: κ·Έλ λ””μ–ΈνΈ λ²΅ν„°λ¥Ό κ³„μ‚°ν•λ” ν•¨μ
grad = gradient(var)

# 1-1) κ·Έλ λ””μ–ΈνΈλ” λ²΅ν„°λ” μ΄λ―€λ΅ μ λ€κ°’ κµ¬ν• μ μ—†μΌλ―€λ΅, λ²΅ν„°μ λ…Έλ¦„(norm)μ„ κ³„μ‚°ν•΄μ„ μΆ…λ£μ΅°κ±΄μ„ μ„¤μ •
while(norm(grad) > eps):
    
    # 1-2) `x β’ Ξ»fβ€²(x)` μ„ κ³„μ‚°ν•λ” λ¶€λ¶„. lr(ν•™μµλ¥ )μ€ ν•™μµλ¥ λ΅μ„, λ―Έλ¶„μ„ ν†µν•΄ μ—…λ°μ΄νΈν•λ” μ†λ„λ¥Ό μ΅°μ ν•λ‹¤.
    var = var - (lr * grad)
    
    # 1-3) μΆ…λ£μ΅°κ±΄μ΄ μ„±λ¦½ν•κΈ° μ „κΉμ§€ 
     grad = gradient(var)
```
